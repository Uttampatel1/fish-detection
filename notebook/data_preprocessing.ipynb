```python
# chlor.loc[chlor["sensor"] == "AQUA", "sensor"] = 1
# chlor.loc[chlor["sensor"] == "TERA", "sensor"] = 2
chl_df['sensor'] = chl_df['sensor'].replace({
                                    'AQUA' : 1,
                                    'TERA' : 2
                                })
```
